```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
new_data_df = pd.read_csv("C:/Users/sadhi/Downloads/worldpopulation.csv",on_bad_lines='skip')
```

```python
print(new_data_df.head())
```

```python
top_20_countries_2022 = new_data_df[['Country Name', '2022']].dropna().sort_values(by='2022', ascending=False).head(20)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(top_20_countries_2022['Country Name'], top_20_countries_2022['2022'], color='red')
plt.title('Top 20 Countries by Population in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=60)
plt.show()
```

```python
population_2022 = new_data_df['2022'].dropna()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(population_2022, bins=30, edgecolor='black')
plt.title('Distribution of Population in 2022')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
country_name = 'United States'
country_data = new_data_df[new_data_df['Country Name'] == country_name].iloc[0, 4:-2].dropna()
```

```python
plt.figure(figsize=(14, 8))
plt.bar(country_data.index, country_data.values, color='yellow')
plt.title(f'Population Growth from 1960 to 2022 for {country_name}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```
